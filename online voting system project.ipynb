```python
# Online Voting System

# Candidate list
candidates = {
    1: "Alice",
    2: "Bob",
    3: "Charlie"
}


votes = {candidate: 0 for candidate in candidates.values()}


voters = set()

def display_candidates():
    print("\n--- Candidate List ---")
    for cid, name in candidates.items():
        print(f"{cid}. {name}")

def cast_vote(voter_id):
    if voter_id in voters:
        print("‚ö†Ô∏è You have already voted. Duplicate voting is not allowed.")
        return
    
    display_candidates()
    try:
        choice = int(input("Enter the candidate number you want to vote for: "))
        if choice in candidates:
            candidate_name = candidates[choice]
            votes[candidate_name] += 1
            voters.add(voter_id)
            print(f"‚úÖ Vote cast successfully for {candidate_name}!")
        else:
            print("‚ùå Invalid choice. Please try again.")
    except ValueError:
        print("‚ùå Invalid input. Please enter a number.")

def display_results():
    print("\n--- Voting Results ---")
    for candidate, count in votes.items():
        print(f"{candidate}: {count} votes")
    # Show winner
    winner = max(votes, key=votes.get)
    print(f"\nüèÜ Current Leading Candidate: {winner}")

def main():
    while True:
        print("\n===== Online Voting System =====")
        print("1. Login and Vote")
        print("2. Display Results")
        print("3. Exit")
        
        choice = input("Enter your choice: ")
        
        if choice == "1":
            voter_id = input("Enter your Voter ID: ")
            cast_vote(voter_id)
        elif choice == "2":
            display_results()
        elif choice == "3":
            print("üëã Exiting... Thank you for participating!")
            break
        else:
            print("‚ùå Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
